```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
Params = [[(5, 5), (5, 5)], [(10, 5), (5, 5)], [(5, 10), (5, 5)], [(10, 10), (5, 5)]]
X_samples = []

for param in Params:
    a = []
    for q in param:
        sample = np.random.normal(q[0], q[1], 100)
        a.append(sample)
    X_samples.append(a)
```

### Проверить гипотезы о нормальности с помощью критерия Шапиро-Уилка, гипотезы о равенстве средних и дисперсий;

```python
# для первого случая, когда Мат.Ожидания и Дисперсии равны

shapiro_stat1, shapiro_p1 = stats.shapiro(X_samples[0][0])
shapiro_stat2, shapiro_p2 = stats.shapiro(X_samples[0][1])

print("Проверка нормальности (Шапиро-Уилка):")
print(f"Выборка 1: статистика = {shapiro_stat1:.4f}, p-value = {shapiro_p1:.4f}")
print(f"Выборка 2: статистика = {shapiro_stat2:.4f}, p-value = {shapiro_p2:.4f}")

# 2. Проверка равенства средних (t-критерий Стьюдента)
t_stat, t_p = stats.ttest_ind(X_samples[0][0], X_samples[0][1], equal_var=True)

print("\nПроверка равенства средних (t-критерий Стьюдента):")
print(f"Статистика = {t_stat:.4f}, p-value = {t_p:.4f}")

# 3. Проверка равенства дисперсий (критерий Левена)
levene_stat, levene_p = stats.levene(X_samples[0][0], X_samples[0][1])

print("\nПроверка равенства дисперсий (критерий Левена):")
print(f"Статистика = {levene_stat:.4f}, p-value = {levene_p:.4f}")
```

```python
# для второго случая, когда Мат.Ожидания существенно отличаются

shapiro_stat1, shapiro_p1 = stats.shapiro(X_samples[1][0])
shapiro_stat2, shapiro_p2 = stats.shapiro(X_samples[1][1])

print("Проверка нормальности (Шапиро-Уилка):")
print(f"Выборка 1: статистика = {shapiro_stat1:.4f}, p-value = {shapiro_p1:.4f}")
print(f"Выборка 2: статистика = {shapiro_stat2:.4f}, p-value = {shapiro_p2:.4f}")

t_stat, t_p = stats.ttest_ind(X_samples[1][0], X_samples[1][1], equal_var=True)

print("\nПроверка равенства средних (t-критерий Стьюдента):")
print(f"Статистика = {t_stat:.4f}, p-value = {t_p:.4f}")

levene_stat, levene_p = stats.levene(X_samples[1][0], X_samples[1][1])
print("\nПроверка равенства дисперсий (критерий Левена):")
print(f"Статистика = {levene_stat:.4f}, p-value = {levene_p:.4f}")
```

```python
# для второго случая, когда Дисперсии существенно отличаются

shapiro_stat1, shapiro_p1 = stats.shapiro(X_samples[2][0])
shapiro_stat2, shapiro_p2 = stats.shapiro(X_samples[2][1])

print("Проверка нормальности (Шапиро-Уилка):")
print(f"Выборка 1: статистика = {shapiro_stat1:.4f}, p-value = {shapiro_p1:.4f}")
print(f"Выборка 2: статистика = {shapiro_stat2:.4f}, p-value = {shapiro_p2:.4f}")

t_stat, t_p = stats.ttest_ind(X_samples[2][0], X_samples[2][1], equal_var=True)

print("\nПроверка равенства средних (t-критерий Стьюдента):")
print(f"Статистика = {t_stat:.4f}, p-value = {t_p:.4f}")

levene_stat, levene_p = stats.levene(X_samples[2][0], X_samples[2][1])

print("\nПроверка равенства дисперсий (критерий Левена):")
print(f"Статистика = {levene_stat:.4f}, p-value = {levene_p:.4f}")
```

```python
# для второго случая, когда Мат.Ожидания и Дисперсии существенно отличаются

shapiro_stat1, shapiro_p1 = stats.shapiro(X_samples[3][0])
shapiro_stat2, shapiro_p2 = stats.shapiro(X_samples[3][1])

print("Проверка нормальности (Шапиро-Уилка):")
print(f"Выборка 1: статистика = {shapiro_stat1:.4f}, p-value = {shapiro_p1:.4f}")
print(f"Выборка 2: статистика = {shapiro_stat2:.4f}, p-value = {shapiro_p2:.4f}")

t_stat, t_p = stats.ttest_ind(X_samples[3][0], X_samples[3][1], equal_var=True)

print("\nПроверка равенства средних (t-критерий Стьюдента):")
print(f"Статистика = {t_stat:.4f}, p-value = {t_p:.4f}")

levene_stat, levene_p = stats.levene(X_samples[3][0], X_samples[3][1])

print("\nПроверка равенства дисперсий (критерий Левена):")
print(f"Статистика = {levene_stat:.4f}, p-value = {levene_p:.4f}")
```

### На одном графике построить гистограммы Х1 и Х2, оценки плотностей для обеих выборок, вертикальными линиями отметить средние значения;

```python
# Оценка плотности
kde1 = stats.gaussian_kde(X_samples[0][0])
kde2 = stats.gaussian_kde(X_samples[0][1])

# Средние значения
mean1 = np.mean(X_samples[0][0])
mean2 = np.mean(X_samples[0][1])

# Построение графика
plt.figure(figsize=(12, 6))

# Гистограммы
sns.histplot(X_samples[0][0], bins=30, color='blue', alpha=0.5, label='Выборка X1', kde=False, stat='density')
sns.histplot(X_samples[0][1], bins=30, color='orange', alpha=0.5, label='Выборка X2', kde=False, stat='density')

# Оценки плотностей
x_vals = np.linspace(min(min(X_samples[0][0]), min(X_samples[0][1])), max(max(X_samples[0][0]), max(X_samples[0][1])), 1000)
plt.plot(x_vals, kde1(x_vals), color='blue', linestyle='--', label='Плотность X1')
plt.plot(x_vals, kde2(x_vals), color='orange', linestyle='--', label='Плотность X2')

# Вертикальные линии для средних
plt.axvline(mean1, color='blue', linestyle=':', linewidth=2, label=f'Среднее X1 = {mean1:.2f}')
plt.axvline(mean2, color='orange', linestyle=':', linewidth=2, label=f'Среднее X2 = {mean2:.2f}')

# Легенда и подписи
plt.title('Гистограммы и оценки плотностей для выборок X1 и X2. Мат.Ожидания и Дисперсии равны.')
plt.xlabel('Значения')
plt.ylabel('Плотность')
plt.legend()

plt.show()
```
